Question 1 : Explain the purpose of version control systems like Git. Why are they essential in modern software development?


Answer : The purpose of Version Control Systems (VCS) like Git is to track and manage changes to files over time, primarily source code, allowing individuals and teams to collaborate effectively, recover from mistakes, and maintain a complete historical record of our project.

VCS is indispensable today due to the complexity, scale, and collaborative nature of modern software projects:
- Team Collaboration : Helps in the team collaboration.
- Support for DevOps and CI/CD : Version control is the foundation of modern continuous integration and continuous delivery (CI/CD) pipelines. Automated builds, tests, and deployments are typically triggered by code commits in the VCS, ensuring a smooth and rapid development lifecycle.
- Agility and Iteration : The ability to easily create "cheap" branches allows for rapid experimentation and the development of multiple features in parallel.
- Code Quality and Review : Platforms built on Git, like GitHub and GitLab, formalize collaboration through Pull Requests (or Merge Requests). These are mechanisms that require peer review of code changes before they are merged into the main codebase, which significantly improves code quality and knowledge sharing.

- Approve and Merge




Question 3: What is the purpose of a .gitignore file in a Git repository? Provide an example of how to ignore all .log files.

Answer :
The primary purpose of a `.gitignore` file is to specify files and directories that Git should intentionally ignore and not track.

This is crucial for excluding files that:

- Are Generated or Temporary: Files created by compilers, build systems, or runtime processes (e.g., compiled code, caches, temporary logs). These files can be easily regenerated and shouldn't be versioned.

- Contain Sensitive Information: Configuration files that hold secrets, API keys, or credentials (e.g., .env files).

- Are Local to the Developer/OS: Files specific to an individual's operating system or IDE configuration (e.g., macOS's .DS_Store, or IDE settings folders like .vscode/).

Example:

The .gitignore file uses pattern matching. To ignore all files with the .log extension, regardless of their subdirectory location, the wildcard (*) is used.

To implement this, one would create a file named .gitignore in the root of the project and add the following line:



```
# This line ignores all files ending with the .log extension,
# anywhere in the repository.
*.log
```
