Answer:



```
const fs = require('fs').promises;

async function readMyFile() {
    try {
        const data = await fs.readFile('data.txt', 'utf8');
        console.log("File Content:");
        console.log(data);
    } catch (err) {
        console.error("Error reading the file:", err.message);
    }
}

readMyFile();
```

Answer:



```

const fs = require('fs').promises;

async function createLogFile() {
    try {
        const content = "Task completed.";
        
        await fs.writeFile('log.txt', content);
        
        console.log("File 'log.txt' has been created successfully!");
    } catch (err) {
        console.error("Error writing to file:", err.message);
    }
}

createLogFile();
```

Answer:



```
// server.js
const http = require('http');

// Create the server
const server = http.createServer((req, res) => {
    // Set the response HTTP header with HTTP status and Content-Type
    res.writeHead(200, { 'Content-Type': 'text/plain' });
    
    // Send the response body "Server is running."
    res.end('Server is running.');
});

// Logic to make the server listen on port 3000
const PORT = 3000;
server.listen(PORT, () => {
    console.log(`Server is successfully running at http://localhost:${PORT}`);
});
```

Answer:


```




const http = require('http');

const server = http.createServer((req, res) => {
    
    res.writeHead(200, { 'Content-Type': 'text/plain' });

    
    if (req.url === '/') {
        res.end('Homepage');
    } else if (req.url === '/about') {
        res.end('About Us');
    } else {
        
        res.writeHead(404, { 'Content-Type': 'text/plain' });
        res.end('404 Not Found');
    }
});

const PORT = 3000;
server.listen(PORT, () => {
    console.log(`Routing server is running at http://localhost:${PORT}`);
});


```

Question 10: Write a script that efficiently reads a large file (large.txt) and outputs its content to the console using streams.

Answer:



```

const fs = require('fs');

const readableStream = fs.createReadStream('large.txt', { encoding: 'utf8' });

readableStream.pipe(process.stdout);

readableStream.on('error', (err) => {
    console.error("\nAn error occurred while reading the file:", err.message);
});

readableStream.on('end', () => {
    console.log("\n\n--- Finished reading file ---");
});
```
