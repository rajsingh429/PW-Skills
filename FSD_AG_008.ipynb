Question 1 : What is React?

Answer:
React is an open-source JavaScript library developed by Meta for building user interfaces, particularly for single-page applications. It allows developers to create reusable UI components that manage their own state. React uses a Virtual DOM to efficiently update the UI by only re-rendering parts of the page that have changed, rather than reloading the entire DOM.

Question 2 : Write a React functional component named `Greeting` that accepts a
`name` prop and displays: "Hello, [name]!".

Answer:


```
const Greeting = (props) => {
  return (
    <div>
      <h1>Hello, {props.name}!</h1>
    </div>
  );
};

export default Greeting;
```



Question 3 : What is JSX and why is it useful in React?

Answer:

JSX (JavaScript XML) is a syntax extension for JavaScript that allows you to write HTML-like structures directly within your JavaScript code.

- Readability: It makes the structure of the UI easier to visualize and understand.

- Performance: JSX is compiled into React.createElement() calls, which React uses to build the Virtual DOM efficiently.

- Safety: It helps prevent Cross-Site Scripting (XSS) attacks by escaping values before rendering them.

Question 4 : Explain how props support React’s one-way data flow

Answer :
In React, data flows in a single direction: from parent to child. Props (short for properties) are the vehicle for this flow.

1.A parent component passes data down to its children via attributes.

2.Props are read-only (immutable) for the child component.

3.This ensures that the child cannot modify the parent's state directly, making the application more predictable and easier to debug.

Question 5 : Create a functional component ClickMeButton that renders a button. It
accepts onClick and label props. When clicked, it executes the passed onClick
handler

Answer :


```
const ClickMeButton = ({ onClick, label }) => {
  return (
    <button onClick={onClick}>
      {label}
    </button>
  );
};

export default ClickMeButton;
```

component FruitList that renders an unordered list of these fruits using .map() and
includes a key.

Answer :




```

const FruitList = () => {
  const fruits = ['Apple', 'Banana', 'Cherry'];

  return (
    <ul>
      {fruits.map((fruit, index) => (
        <li key={index}>{fruit}</li>
      ))}
    </ul>
  );
};

```

Question 7 : Why is the key attribute important when rendering lists in React? What
goes wrong if keys are missing or misuse array index?

Answr :

Keys help React identify which items in a list have changed, been added, or been removed.

- Importance: They are essential for the reconciliation process, allowing React to re-order elements rather than re-rendering the entire list.

- Missing Keys: If keys are missing, React might lose track of component state (like focus or input values) when the list changes.

- Misusing Indices: Using the array index as a key is risky if the list order can change (via sorting or deleting). This can lead to UI bugs where the wrong data stays associated with the wrong DOM element.

Question 8 : Write a component UserList that receives a users prop—an array of
objects each with id and name. Render each user’s name in a <p> tag, using map()
and key.

Answer:

Render each user's name in a` <p>` tag, using map() and key.


```
const UserList = ({ users }) => {
  return (
    <div>
      {users.map((user) => (
        <p key={user.id}>{user.name}</p>
      ))}
    </div>
  );
};
```

a NumberList child component for rendering, and passes a click handler function to
log the clicked number. Show one-way data flow and event handling.

Answer:



```
import React from 'react';

const NumberList = ({ numbers, onNumberClick }) => (
  <ul>
    {numbers.map((num) => (
      <li key={num} onClick={() => onNumberClick(num)} style={{ cursor: 'pointer' }}>
        Number: {num}
      </li>
    ))}
  </ul>
);

const App = () => {
  const numbers = [10, 20, 30, 40, 50];

  const handleNumberClick = (num) => {
    console.log("Clicked number:", num);
  };

  return (
    <div>
      <h1>Number Viewer</h1>
      <NumberList numbers={numbers} onNumberClick={handleNumberClick} />
    </div>
  );
};

export default App;
```



Question 10 : Design a component ProductList that receives products (array of
objects with id, name, price). Render a list of product names and prices using
.map(). Each item when clicked should call a prop onProductSelect(product).
Include the proper key.

Answer:



```
const ProductList = ({ products, onProductSelect }) => {
  return (
    <div className="product-list">
      {products.map((product) => (
        <div
          key={product.id}
          onClick={() => onProductSelect(product)}
          style={{ border: '1px solid #ccc', margin: '10px', padding: '10px', cursor: 'pointer' }}
        >
          <h3>{product.name}</h3>
          <p>Price: ${product.price}</p>
        </div>
      ))}
    </div>
  );
};
```
