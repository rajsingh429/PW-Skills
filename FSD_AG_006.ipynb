Question 1 : Explain what the forEach method does in JavaScript arrays and write code
that uses forEach to display each student's name from an array of student objects to the
console.

```
// Array of student objects
const students = [
{ name: "Rahul", age: 20 },
{ name: "Priya", age: 19 },
{ name: "Amit", age: 21 },
{ name: "Sneha", age: 20 }
];
```
Answer :
The Array.prototype.forEach() Method

The forEach() method is a built-in function in JavaScript used specifically for iterating over the elements of an array.

What it does:

It executes a provided callback function once for each element in the array, in ascending order of index.



```
const students = [
    { name: "Rahul", age: 20 },
    { name: "Priya", age: 19 },
    { name: "Amit", age: 21 },
    { name: "Sneha", age: 20 }
];

console.log("--- Student Names Log ---");

students.forEach((student, index) => {
    
    console.log(`Index ${index}: ${student.name}`);
});
```

array of employee objects by their salary in descending order (highest salary first).



```
// Array of employee objects
const employees = [
 { name: "Rajesh", salary: 45000 },
 { name: "Priya", salary: 55000 },
 { name: "Amit", salary: 40000 },
 { name: "Kavya", salary: 60000 }
];
```



Answer:The sort() method in JavaScript is used to sort the elements of an array in place and returns the sorted array. It is important to note that because the sort is performed in place, the original array is modified.

When sorting an array of objects, the compareFunction receives the objects themselves (a and b). You then access the property you want to sort by (e.g., a.salary and b.salary) and perform the comparison calculation.

Ascending (Lowest to Highest): a.salary - b.salary

Descending (Highest to Lowest): b.salary - a.salary



```
const employees = [
    { name: "Rajesh", salary: 45000 },
    { name: "Priya", salary: 55000 },
    { name: "Amit", salary: 40000 },
    { name: "Kavya", salary: 60000 }
];

const sortBySalaryDescending = (a, b) => {
    return b.salary - a.salary;
};

employees.sort(sortBySalaryDescending);

console.log("Employees sorted by salary in descending order (Highest to Lowest):");
console.log(employees);


```

in a shopping cart. Each item has a price and quantity. Also find the most expensive
single item in the cart.



```
// Shopping cart with items
const cart = [
{ name: "Laptop", price: 50000, quantity: 1 },
{ name: "Mouse", price: 500, quantity: 2 },
{ name: "Keyboard", price: 1500, quantity: 1 },
{ name: "Monitor", price: 15000, quantity: 1 }
];
```



Answer:



```
const cart = [
    { name: "Laptop", price: 50000, quantity: 1 },
    { name: "Mouse", price: 500, quantity: 2 },
    { name: "Keyboard", price: 1500, quantity: 1 },
    { name: "Monitor", price: 15000, quantity: 1 },
    { name: "External SSD", price: 8000, quantity: 3 }
];


const analysisResult = cart.reduce((accumulator, item) => {
    // 1. Calculate the total value
    const itemTotal = item.price * item.quantity;
    accumulator.totalValue += itemTotal;

  
    if (item.price > accumulator.mostExpensiveItem.price) {
        accumulator.mostExpensiveItem = {
            name: item.name,
            price: item.price
        };
    }

    return accumulator;
}, {
   
    totalValue: 0,
    mostExpensiveItem: {
        name: "",
        price: 0
    }
});


console.log("--- Shopping Cart Analysis ---");
console.log(`Total Value of all items: ${analysisResult.totalValue}`);
console.log(`Most Expensive Single Item: ${analysisResult.mostExpensiveItem.name} (Price: ${analysisResult.mostExpensiveItem.price})`);

/*
Expected Output:
Total Value of all items: 84500 (50000*1 + 500*2 + 1500*1 + 15000*1 + 8000*3)
Most Expensive Single Item: Laptop (Price: 50000)
*/
```

two colors from an array of colors into separate variables using destructuring.



```
// Array of colors
const colors = ["red", "blue", "green", "yellow", "purple"];
```

Answer:
Array destructuring is a powerful syntax in JavaScript that makes it possible to unpack values from arrays (or properties from objects) into distinct variables. It provides a more concise and readable way to assign variables compared to accessing elements one by one using their index.

Key Concepts

- Extraction: You define the variables you want to extract on the left side of the assignment operator, wrapped in square brackets ([]).

- Assignment: The array being destructured is placed on the right side.

- Positional Mapping: Variables are assigned values based on their position in the array. The first variable gets the value of the first element, the second variable gets the value of the second element, and so on.



```
const colors = ["red", "blue", "green", "yellow", "purple"];

const [firstColor, secondColor] = colors;
```

prices as arguments and returns their sum. Use the rest operator to handle the variable
number of parameters.



Answer:



```
const calculateTotal = (...prices) => {
  return prices.reduce((acc, currentPrice) => acc + currentPrice, 0);
};


const total1 = calculateTotal(10.50, 25.00, 5.99);
console.log(`Total 1 (10.50, 25.00, 5.99): $${total1.toFixed(2)}`);


const total2 = calculateTotal(100, 50, 200, 75, 10);
console.log(`Total 2 (100, 50, 200, 75, 10): $${total2.toFixed(2)}`);

const total3 = calculateTotal();
console.log(`Total 3 (no arguments): $${total3.toFixed(2)}`);


const shoppingCart = [3.49, 12.99, 45.00];
const cartTotal = calculateTotal(...shoppingCart);
console.log(`Cart Total (using spread operator on array): $${cartTotal.toFixed(2)}`);
```

that contains all students from both arrays. Also, you have two objects representing a
student's basic info and marks. Write code using the spread operator to solve both
scenarios.



```
const classA = ["Rahul", "Priya", "Amit"];
const classB = ["Sneha", "Vikram", "Pooja"];
```



Answer:



```
const classA = ["Rahul", "Priya", "Amit"];
const classB = ["Sneha", "Vikram", "Pooja"];

const allStudents = [...classA, ...classB];

console.log("--- Scenario 1: Array Merging ---");
console.log("Class A:", classA);
console.log("Class B:", classB);
console.log("All Students (Merged):", allStudents);

const studentInfo = {
    id: 101,
    name: "Rahul Sharma",
    class: "10th Grade"
};

const studentMarks = {
    math: 95,
    science: 92,
    english: 88,
    average: 91.67
};


const studentProfile = {
    ...studentInfo,
    ...studentMarks,
    // You can also add new properties here:
    status: "Excellent"
};

console.log("\n--- Scenario 2: Object Merging ---");
console.log("Student Info:", studentInfo);
console.log("Student Marks:", studentMarks);
console.log("Student Profile (Merged):", studentProfile);
```
