Question 1:  What is Pug? Why might developers prefer Pug over writing raw HTML for
server-rendered pages?

Answer:

Pug (formerly known as Jade) is a high-performance template engine for Node.js that compiles into HTML. It is specifically designed to simplify the process of creating server-rendered pages by providing a more streamlined syntax than traditional HTML.

Developers often prefer Pug over writing raw HTML for several key reasons:


- Conciseness and Readability: Pug uses an indentation-based syntax and eliminates the need for closing tags (like </div> or </span>), which significantly reduces the total lines of code and makes the structure easier to scan visually.


- Dynamic Logic: It allows developers to write inline JavaScript directly within the template, making it simple to handle loops, conditionals, and variables without switching back and forth between different file types.


- Code Reusability: Pug supports "includes" and "inheritance" (layouts), which allow developers to create a single base template and reuse it across multiple pages, ensuring consistency and reducing manual updates.


- High Performance: Because it is compiled into efficient JavaScript functions, it renders pages very quickly on the server side.


Question 2:  Initialize an Express server that uses Pug as the view engine. Create a route
`/welcome` that renders a Pug view `welcome.pug`, passing a variable `name`. The view should
greet the user by name. Show code for server setup and the Pug template.

Answer:  

Server Setup (app.js)

```
const express = require('express');
const app = express();
const port = 3000;

app.set('view engine', 'pug');

app.get('/welcome', (req, res) => {

    const userName = req.query.name || "Guest";
    res.render('welcome', { name: userName });
});

app.listen(port, () => {
    console.log(`Server running at http://localhost:${port}`);
});
```

Pug Template (views/welcome.pug)



Question 3: Explain what a RESTful API is. Then list four HTTP methods (GET, POST,
PUT/PATCH, DELETE) and describe what each should do in a REST design for a resource
“/tasks”.

Answer :
A RESTful API (Representational State Transfer) is an architectural style for designing networked applications that allows different software systems to communicate over HTTP. It treats data as "resources" (like tasks, users, or products) that can be accessed and manipulated using standard web protocols

the four primary HTTP methods function as follows:


- GET: Used to retrieve data from the server. In this context, it would fetch a list of all tasks or a specific task if an ID is provided (e.g., /tasks/1).


- POST: Used to send data to the server to create a new resource. For "/tasks", this would involve submitting details for a brand-new task to be added to the collection.


- PUT/PATCH: Used to update an existing resource. PUT typically replaces the entire task with new data, while PATCH is used for partial updates (like just changing a task's status from "pending" to "completed").


- DELETE: Used to remove a specific resource from the server. For "/tasks", it would permanently delete a task identified by its unique ID.

Question 4: Using Express, define routes for a Task resource: GET /tasks, POST /tasks,
PUT /tasks/:id, DELETE /tasks/:id. In each handler, just send JSON saying which
operation was called (no DB required yet). Write the Express route definitions.

Answer:  

```
const express = require('express');
const app = express();


app.get('/tasks', (req, res) => {
    res.json({ message: "Operation: GET /tasks (Fetch all tasks)" });
});

app.post('/tasks', (req, res) => {
    res.json({ message: "Operation: POST /tasks (Create a task)" });
});

app.put('/tasks/:id', (req, res) => {
    const taskId = req.params.id;
    res.json({ message: `Operation: PUT /tasks/${taskId} (Update task)` });
});

app.delete('/tasks/:id', (req, res) => {
    const taskId = req.params.id;
    res.json({ message: `Operation: DELETE /tasks/${taskId} (Delete task)` });
});

app.listen(3000, () => console.log('Task API running on port 3000'));
```

using local strategy: user submits credentials, server checks them, and if valid, logs in (you can
assume a dummy password check). You don’t need to provide full code, just outline key steps
(Express + Passport + session or token).

Answer:

Passport.js is a flexible authentication middleware for Node.js. Its primary job is to handle the complex logic of logging users in—supporting various methods like traditional passwords, Google, or Facebook—while keeping the process modular and easy to integrate into Express applications.

Flow of User Login (Local Strategy)

- Credential Submission: The user enters their username and password into a login form, which is sent to the server via a POST request.


- Strategy Trigger: Express receives the request and passes it to Passport’s local strategy.

- Verification (Dummy Check): The server looks up the user in the database. It compares the submitted password with the stored one (in this case, a simple "dummy" match).


- Serialization: If the credentials are valid, Passport calls serializeUser to determine which data (usually the User ID) should be stored in the session cookie.


- Session Establishment: A session is created, and a cookie is sent back to the user's browser to keep them logged in for subsequent requests.


- Success Response: The server redirects the user to a "dashboard" or sends a success JSON message.

a small explanation AND describe a safe practice to avoid this risk when accepting user input
for queries.

Answer :

NoSQL Injection is a vulnerability where an attacker provides malicious input to manipulate the structure or logic of a database query. In a MongoDB + Express application, this typically occurs when user-provided objects (like dictionaries containing MongoDB operators) are passed directly into a query filter without being sanitized

Why it is a Threat

- Data Leakage: Attackers can use operators like $gt (greater than) or $ne (not equal) to bypass authentication or extract sensitive records.

- Bypassing Logic: For example, an attacker could input {"$gt": ""} into a password field. Since most passwords are "greater than" an empty string, the query returns true, effectively logging the attacker in without a valid password.


Safe Practice: Sanitization and Validation
The most effective way to prevent this risk is to ensure that user input is treated strictly as data, never as part of the query command.

Question 9: Your API serves a frontend hosted on another domain. What is CORS and why do
you need to configure it? Show minimal Express code to allow CORS only from
https://myfrontend.com.

Answer:

CORS (Cross-Origin Resource Sharing) is a security mechanism implemented by web browsers to restrict how resources on a web page can be requested from another domain outside the domain from which the first resource was served. By default, browsers block these "cross-origin" requests to prevent malicious websites from reading sensitive data from another site.

Minimal Express Code for CORS Configuration



```
const express = require('express');
const cors = require('cors');
const app = express();

// Configure CORS to allow only a specific domain
const corsOptions = {
  origin: 'https://myfrontend.com'
};

app.use(cors(corsOptions)); // Apply the middleware

app.get('/data', (req, res) => {
  res.json({ message: "This data is accessible to myfrontend.com" });
});

app.listen(3000);
```

middleware requireAdmin that only allows access if role is 'admin', else returns 403. Show
how to use it for route DELETE /admin/delete-user/:id.

Answer:  
Middleware Implementation

```
const requireAdmin = (req, res, next) => {

    if (req.user && req.user.role === 'admin') {
        next(); // User is admin, proceed to the next handler
    } else {

        res.status(403).json({ error: "Forbidden: Admin access required" });
    }
};
```

Route Usage
```

app.delete('/admin/delete-user/:id', requireAdmin, (req, res) => {
    const targetUserId = req.params.id;
    res.json({ message: `Successfully deleted user with ID: ${targetUserId}` });
});
```
