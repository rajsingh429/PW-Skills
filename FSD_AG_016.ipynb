Question 1:  What is a RESTful API? Explain four core principles of RESTful API design and
why they are important.

Answer:  A RESTful API (Representational State Transfer) is an architectural style for designing networked applications that uses HTTP requests to manage data.

Four Core Principles:

- Statelessness: Each request from a client must contain all the information needed to understand and process the request. The server does not store any client context between requests, which improves scalability.

- Client-Server Separation: The client (user interface) and the server (data storage/logic) are independent. This allows them to evolve separately as long as the interface remains consistent.


- Uniform Interface: By using standard HTTP verbs (GET, POST, PUT, DELETE) and resource URIs, the API becomes predictable and easier to use for different clients.


- Layered System: A client cannot tell whether it is connected directly to the end server or to an intermediary (like a load balancer or proxy), which enhances security and performance.

Question 2:  Write Express.js routes for a users resource supporting create, read (all and by
id), update, and delete operations.

Answer:  



```
const express = require('express');
const router = express.Router();

router.post('/users', (req, res) => {
   
    res.status(201).send("User created successfully");
});

router.get('/users', (req, res) => {
    
    res.status(200).json({ message: "Fetched all users" });
});

router.get('/users/:id', (req, res) => {
    const userId = req.params.id;
   
    res.status(200).json({ message: `Fetched user with ID: ${userId}` });
});

router.put('/users/:id', (req, res) => {
    const userId = req.params.id;
    // Logic to update user details in the database using req.body
    res.status(200).send(`User ${userId} updated`);
});

router.delete('/users/:id', (req, res) => {
    const userId = req.params.id;
   
    res.status(200).send(`User ${userId} deleted`);
});

module.exports = router;
```

username and password.

Answer:  

Passport.js is a flexible, modular authentication middleware for Node.js that simplifies the process of managing complex login logic. Its primary role is to provide a consistent interface for authenticating requests through various "strategies," such as local databases or third-party providers like Google.

The Role of Passport.js

- Abstraction: It decouples the authentication logic from the application routes, allowing developers to swap authentication methods without changing the core application structure.


- Middleware Integration: It integrates seamlessly into the Express.js request-response cycle, attaching the authenticated user object to req.user for easy access in subsequent routes.


- Session Management: It handles the serialization and deserialization of user data to maintain a persistent login state via sessions.

Login Flow: Username and Password

- User Submission: The user enters their credentials into a login form and submits a POST request to the server.


- Strategy Trigger: The application calls passport.authenticate('local'), which triggers the LocalStrategy.


- Database Verification: The strategy's callback function queries the database to find the user by their username and compares the submitted password with the hashed password stored in the database.

- Verification Result:


- Success: If the credentials match, the strategy calls done(null, user).


- Failure: If the credentials are invalid, it calls done(null, false) with an optional error message.


- Serialization: Upon success, Passport calls passport.serializeUser() to determine which data (usually the User ID) should be stored in the session cookie.


- Session Establishment: A session is created, and the user is redirected to the protected area of the application.

Question 4: Explain the OAuth login flow using Google in an Express application.

Answer:

The OAuth 2.0 login flow allows your Express application to delegate authentication to Google, ensuring you never have to handle or store user passwords directly.

The Step-by-Step Flow

- User Initiation: The user clicks a "Login with Google" button on your website, which sends a request to your Express server.


- Redirection to Google: Your server redirects the user to Google’s OAuth 2.0 server with your Client ID and requested Scopes (e.g., email and profile).


- User Consent: Google displays a consent screen to the user, asking if they allow your application to access their information.


- Authorization Code: Once the user grants permission, Google redirects them back to your application’s Callback URL (e.g., /auth/google/callback) with a temporary Authorization Code.

- Token Exchange: Your Express server receives this code and sends it back to Google's token endpoint along with your Client Secret. This proves that the request is coming from your authorized server.


- Access Token Issuance: Google verifies the code and secret, then issues an Access Token (and often an ID Token) back to your server.

- Final Authentication: Your application uses the Access Token to fetch the user's profile information from Google. You then log the user into your system by creating a session or issuing a JWT.

Why it is used

- Security: You don't store passwords, reducing the risk of data breaches.


- User Experience: Users don't need to create a new account; they can use their existing Google credentials.


- Scoped Access: You only get the specific data (like email) that the user approves.

Question 5: Compare session-based authentication and token-based (JWT) authentication.
When would you use each?

Answer:

Comparison of Authentication Methods

- Session-based Authentication
This is a stateful approach where the server is responsible for keeping track of the user's login status. When a user logs in, the server creates a session record in its memory or a database and sends a unique Session ID to the client via a cookie. On every subsequent request, the browser sends this cookie back, and the server verifies it against its stored data. This makes it very easy to revoke access instantly by deleting the session from the server, but it becomes harder to scale across multiple servers because they all need access to the same session data.

- Token-based (JWT) Authentication
This is a stateless approach where the server does not store any information about the user's session. Instead, after a successful login, the server generates a JSON Web Token (JWT) and sends it to the client. The client stores this token (usually in Local Storage) and sends it in the header of every request. Since the token itself contains all the necessary user information and is digitally signed, the server can verify it without checking a database. While this makes scaling very easy, revoking a token before it expires is much more complex than killing a session.


When to Use Each

- Use Session-based Authentication if you are building a traditional, monolithic web application where the frontend and backend are on the same domain and you need total control over active user sessions.


- Use Token-based (JWT) Authentication for modern Single Page Applications (SPAs), mobile apps, or microservice architectures where you need the API to be stateless and highly scalable across different domains

technique for Node.js applications.

Answer:

SQL Injection (SQLi)

SQL injection occurs when an attacker inserts malicious SQL code into an input field or API request. If the application does not properly sanitize this input, the malicious code is executed by the backend database. This can allow attackers to bypass authentication, view private user data, or even delete entire database tables. For example, a malicious user might enter ' OR '1'='1 into a login field to trick the database into granting access without a valid password.

NoSQL Injection

Similar to SQL injection, NoSQL injection targets non-relational databases like MongoDB. Instead of using SQL commands, attackers use database-specific operators—often sent via JSON—to manipulate queries. For instance, sending a request like {"username": {"$gt": ""}} might trick the database into returning all users where the username is "greater than" an empty string, effectively leaking the entire user list.

Practical Prevention Technique:

 Parameterized Queries (Input Sanitization)
The most effective practical prevention technique for Node.js applications is the use of Parameterized Queries or Object-Relational Mapping (ORM/ODM) libraries.

In a Node.js environment, instead of manually concatenating strings to build a query, we use placeholders. When using an ODM like Mongoose (for MongoDB) or an ORM like Sequelize (for SQL), the library automatically treats the input as data rather than executable code.


Question 8: What is CORS and why must it be configured in APIs? Write Express.js code to
allow requests only from https://client.app.

Answer:

CORS (Cross-Origin Resource Sharing) is a security mechanism implemented by web browsers that restricts how a resource on one domain can be requested by another domain. By default, browsers block "cross-origin" requests (e.g., a frontend at https://client.app trying to fetch data from an API at https://api.app) to prevent malicious websites from reading sensitive data from other sites.

Why it must be configured:


- Security: It prevents unauthorized websites from making requests to your API on behalf of a user.


- Access Control: It allows you to explicitly "whitelist" trusted domains (like your own frontend application) so they can interact with your backend services.


- Preflight Requests: It enables the browser to perform a "preflight" check (using the OPTIONS method) to ensure the server understands and permits the specific HTTP methods or headers being used.





```
const express = require('express');
const cors = require('cors');
const app = express();

const corsOptions = {
    origin: 'https://client.app',
    optionsSuccessStatus: 200     
};

app.use(cors(corsOptions));

app.get('/data', (req, res) => {
    res.json({ message: "This data is accessible only by https://client.app" });
});

app.listen(3000, () => console.log('Server running on port 3000'));
```

using the App Router.

Answer:  

Next.js is an open-source web development framework built on top of the React library. While React is a library focused on building user interfaces, Next.js is a full-featured framework that provides the tools and structure needed to build production-ready, high-performance applications.

Why Next.js is used over plain React:


- Rendering Options: Unlike plain React, which primarily uses Client-Side Rendering (CSR), Next.js offers Server-Side Rendering (SSR) and Static Site Generation (SSG). This allows pages to be pre-rendered on the server, leading to faster initial load times and significantly better SEO.


- Built-in Optimization: Next.js includes automatic optimizations for images, fonts, and scripts that you would otherwise have to configure manually in a plain React setup.


- Full-Stack Capabilities: Next.js allows you to write backend code (API routes) within the same project, eliminating the need for a separate Express server for simple backend logic.


- Zero Configuration: It comes with a built-in compiler and bundler, so features like code splitting and hot module replacement work out of the box without complex Webpack setups.

File-based Routing (App Router)

In Next.js, routing is handled automatically based on the file system structure within a special directory called app. You do not need a third-party library like React Router.

How it works:


- Folders define Routes: Each folder inside the app directory represents a URL segment. For example, a folder named dashboard corresponds to the /dashboard path.


- The page.js file: For a route to be publicly accessible, it must contain a file exactly named page.js (or .tsx). This file defines the UI unique to that path.


- Nested Routing: You can nest folders within folders to create deep URLs. For example, app/shop/products/page.js will map to /shop/products.


- Special Files: The App Router also uses reserved filenames to handle specific behaviors:


- layout.js: Used to create shared UI (like navbars) across multiple pages.


- loading.js: Automatically displays a loading state while the page content is fetching.


- error.js: Handles runtime errors for a specific segment of the application.

Question 10: Explain dynamic routes, middleware, and one performance feature (Suspense or
lazy loading) in Next.js.

Answer:

Dynamic Routes

Dynamic routes are used when you don’t know the exact segment names ahead of time and want to create routes from dynamic data. In the App Router, you create a dynamic route by wrapping a folder's name in square brackets, such as [id] or [slug]. For example, a folder named app/blog/[slug]/page.js would allow you to handle URLs like /blog/hello-world or /blog/nextjs-tips, where the actual value of "slug" is passed as a prop to the page component.

Middleware

Middleware in Next.js allows you to run code before a request is completed. It functions as a gatekeeper, intercepting requests to perform tasks such as authentication checks, bot detection, or redirection. For instance, you can use middleware to check if a user is logged in before allowing them to access the /dashboard route; if they aren't, the middleware can redirect them to the login page automatically.

Performance Feature: Suspense

Suspense is a powerful performance feature in Next.js that allows you to display a fallback UI (like a loading spinner or a skeleton screen) while a specific component is still fetching data.


- Improved Perceived Performance: Instead of waiting for the entire page to load, the user can see parts of the page immediately while the heavier data-fetching components load in the background.


- Selective Hydration: It helps in breaking down the page into smaller chunks, allowing the browser to prioritize interactive elements.
