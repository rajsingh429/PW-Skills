and returns:

● "Eligible" if age is 18 or above

● "Not Eligible" if age is below 18

Use an if-else statement


Answer :

this will be the required code

```

function checkEligibility(age) {
    // We use an if-else statement to check the condition.
    if (age >= 18) {
        return "Eligible";
    } else {
        return "Not Eligible";
    }
}


const age1 = 25;
const age2 = 18;
const age3 = 17;

console.log(`Age ${age1}: ${checkEligibility(age1)}`); // Expected: Eligible
console.log(`Age ${age2}: ${checkEligibility(age2)}`); // Expected: Eligible
console.log(`Age ${age3}: ${checkEligibility(age3)}`); // Expected: Not Eligible


```

corresponding day of the week. Use a switch statement.

Answer :

```

function getDayName(dayNumber) {
    let dayName;

    switch (dayNumber) {
        case 1:
            dayName = "Sunday";
            break;
        case 2:
            dayName = "Monday";
            break;
        case 3:
            dayName = "Tuesday";
            break;
        case 4:
            dayName = "Wednesday";
            break;
        case 5:
            dayName = "Thursday";
            break;
        case 6:
            dayName = "Friday";
            break;
        case 7:
            dayName = "Saturday";
            break;
        default:
            dayName = "Invalid Day Number";
    }

    return dayName;
}


const day1 = 1;
const day5 = 5;
const invalidDay = 10;

console.log(`Day ${day1} is: ${getDayName(day1)}`);
console.log(`Day ${day5} is: ${getDayName(day5)}`);
console.log(`Day ${invalidDay} is: ${getDayName(invalidDay)}`);


```

1. Use a for loop to print each number.
2. Use a while loop to print each number.
3. Use a for-in loop to print each index.
4. Use a for-of loop to print each number.


Answer :

Here is the code for each

```
const numbers = [1, 2, 3, 4, 5];
```

- Using a for loop (Print each number)



```
console.log("1. For Loop (Numbers):");
for (let i = 0; i < numbers.length; i++) {
    console.log(numbers[i]);
}
```

- Using a while loop (Print each number)



```
console.log("\n2. While Loop (Numbers):");
let i = 0;
while (i < numbers.length) {
    console.log(numbers[i]);
    i++;
}
```

- Using a for-in loop (Print each index)



```
console.log("\n3. For-In Loop (Indices):");
for (const index in numbers) {
    console.log(index);
}
```

- Using a for-of loop (Print each number)



```
console.log("\n4. For-Of Loop (Numbers):");
for (const number of numbers) {
    console.log(number);
}
```

function that adds two numbers.

Answer :

Functions in JavaScript :

A function in JavaScript is a block of reusable code designed to perform a particular task. Think of it as a mini-program within a larger program.

Key Concept of the function in js is :

- Reusability: Functions allow you to write a block of code once and execute it multiple times, which reduces repetition (the "Don't Repeat Yourself," or DRY principle).

- Encapsulation: They group a set of statements together to perform a single operation, making the code organized and easier to understand.

- Parameters and Arguments: Functions can accept input values called parameters. When the function is called, the actual values passed to it are called arguments.

- Return Value: A function can optionally process data and return a result using the return keyword. If a function doesn't explicitly return a value, it returns undefined.



Example :

```

function addNumbers(num1, num2) {
    let sum = num1 + num2;
    
    return sum;
}

const result1 = addNumbers(5, 10);
const result2 = addNumbers(20, 3);

console.log(result1);
console.log(result2);
```

greet function should call the callback with the name.


Answer :



```
function greet(name, callback) {
    console.log(`--- Executing greet function for ${name} ---`);
     callback(name);
}

function formalGreeting(personName) {
    console.log(`Hello, ${personName}. It is a pleasure to meet you.`);
}

function casualGreeting(personName) {
    console.log(`Yo, ${personName}! What's up?`);
}


console.log("Scenario 1: Using the formalGreeting callback");
greet("Alice", formalGreeting);


console.log("\nScenario 2: Using the casualGreeting callback");
greet("Bob", casualGreeting);


console.log("\nScenario 3: Using an anonymous arrow function as a callback");
greet("Charlie", (n) => {
    const capitalized = n.toUpperCase();
    console.log(`Callback executed: GREETINGS, ${capitalized}!`);
});
```

1. Trim the whitespace.
2. Convert to uppercase.
3. Check if it includes "World"

Answer :





```
const text = " Hello World! ";
```



1. Trim the Whitespace

```
const trimmedText = text.trim();
console.log(`1. Trimmed Text: "${trimmedText}"`);
// Output: "Hello World!"
```

2. Convert to Uppercase


```
// We can use the trimmed string for a cleaner result
const upperCaseText = trimmedText.toUpperCase();
console.log(`2. Uppercase Text: "${upperCaseText}"`);
// Output: "HELLO WORLD!"
```


3. Check if it includes "World"
```
const includesWorld = trimmedText.includes("World");
console.log(`3. Includes "World" (Case-Sensitive): ${includesWorld}`);
// Output: true
```

1. Add 6 to the end.
2. Remove the first element.
3. Find the index of 4.
4. Check if all numbers are greater than 0
Answer :

let numbers = [1, 2, 3, 4, 5];

1. Add 6 to the end



```
numbers.push(6);
console.log("1. After push(6):", numbers);
// Output: [1, 2, 3, 4, 5, 6]
```

2. Remove the first element



```
numbers.shift();
console.log("2. After shift():", numbers);
// Output: [2, 3, 4, 5, 6]
```
3. Find the index of 4


```
const indexOfFour = numbers.indexOf(4);
console.log("3. Index of 4:", indexOfFour);
// Output: 2 (since 4 is at index 2 in the current array [2, 3, 4, 5, 6])
```

4. Check if all numbers are greater than 0



```
const allGreaterThanZero = numbers.every(number => number > 0);
console.log("4. All numbers > 0:", allGreaterThanZero);
// Output: true
```
