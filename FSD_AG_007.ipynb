Question 01 : Write code that creates a new paragraph element with the text "Welcome
to our website!" and adds it to a div with the id "content" on the webpage. Explain what
createElement and appendChild methods do.

Answer :

```
const newPara = document.createElement("p");

newPara.textContent = "Welcome to our website!";

const contentDiv = document.getElementById("content");
contentDiv.appendChild(newPara);
```

Write code that finds a button with id "myButton" and changes its background color to
blue and text color to white using the element.style property.

Answer:


```
const btn = document.getElementById("myButton");

btn.addEventListener("mouseover", () => {
    btn.style.backgroundColor = "blue";
    btn.style.color = "white";
});


btn.addEventListener("mouseout", () => {
    btn.style.backgroundColor = "";
    btn.style.color = "";
});
```

"john@example.com" and returns the domain part ("example.com") using only
string methods. Do not use regular expressions.

Answer:


```
function extractDomain(email) {
    const atIndex = email.indexOf("@");
    return email.slice(atIndex + 1);
}

// Example usage:
console.log(extractDomain("raj@example.com")); // Output: "example.com"
```

a number, and returns the message repeated that many times, separated by a space.
Do not use loops.

Example:
repeatMessage("hello", 3) should return "hello hello hello"

Answer:


```
function repeatMessage(message, n) {
    return Array(n).fill(message).join(" ");
}

// Example usage:
console.log(repeatMessage("hello", 3)); // Output: "hello hello hello" [cite: 34]
```

following code. Also explain why the output appears in that specific order.

```
console.log('Start');
setTimeout(function() {
console.log('Timeout 1');
}, 0);
console.log('Middle');
setTimeout(function() {
console.log('Timeout 2');
}, 0);
console.log('End');

```

Answer:
Predicted Output:
Start

Middle

End

Timeout 1

Timeout 2

Explanation:

JavaScript is single-threaded. Synchronous code (console.log) runs immediately on the Call Stack. Functions like setTimeout are sent to the Web APIs container. Even with a delay of 0, the callback is placed in the Task Queue. The Event Loop only pushes tasks from the queue to the stack once the stack is completely empty. Thus, all main script logs finish before the timeouts execute.



Question 7 : Write a JavaScript function named startCountdown that logs “Time left:
X” every second, starting from 5 down to 1, and finally logs “Time’s up!”. Use
setInterval and clearInterval only.

Answer:


```
function startCountdown() {
    let count = 5;
    const intervalId = setInterval(() => {
        if (count > 0) {
            console.log(`Time left: ${count}`);
            count--;
        } else {
            console.log("Time's up!");
            clearInterval(intervalId);
        }
    }, 1000);
}

startCountdown();
```

```
function getUserDataAndPosts(userId) {
return fetchUserData(userId)
.then(userData => {
return fetchUserPosts(userData.id)
.then(posts => {
return { user: userData, posts: posts };
});
})
.catch(error => {
console.error('Error:', error);
throw error;
});
}

```

Answer:



```
async function getUserDataAndPosts(userId) {
    try {
        const userData = await fetchUserData(userId);
        const posts = await fetchUserPosts(userData.id);
        return { user: userData, posts: posts };
    } catch (error) {
        console.error('Error:', error);
        throw error;
    }
}
```

Display the weather information in a user-friendly format. Assume you have the
following configuration for the API.

```
const apiKey = 'abcd1234';
const apiUrl =
`https://api.openweathermap.org/data/2.5/weather?q=Mumbai&appid=${apiK
ey}&units=metric`;

```



Answer :

```
const apiKey = 'abcd1234'; // [cite: 80]
const city = 'Bhopal';
const apiUrl = `https://api.openweathermap.org/data/2.5/weather?q=${city}&appid=${apiKey}&units=metric`;

async function getWeather() {
    try {
        const response = await fetch(apiUrl);
        if (!response.ok) throw new Error("Weather data not found");
        
        const data = await response.json();
        console.log(`Current weather in ${data.name}:`);
        console.log(`Temperature: ${data.main.temp}°C`);
        console.log(`Condition: ${data.weather[0].description}`);
    } catch (error) {
        console.error("Fetch Error:", error);
    }
}

getWeather();
```
