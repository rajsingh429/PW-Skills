Answer : Here is complete code for the question :



```
<!DOCTYPE html>
<html>
<head>
    <title>CSS</title>
    <style>
        .grid-container {
            /* Enables CSS Grid layout */
            display: grid;
            
            /* Defines 2 equal columns */
            grid-template-columns: 1fr 1fr;
            
            /* Defines 2 rows (auto-sized) */
            grid-template-rows: auto auto;
            
            /* Adds some spacing between grid cells */
            gap: 10px;
            
            /* Optional: For visual demonstration */
            width: 80%;
            margin: 20px auto;
            padding: 10px;
            border: 3px solid #004d40; /* Dark Teal Border */
            font-family: Arial, sans-serif;
        }

        .grid-item {
            /* Optional: For visual demonstration */
            background-color: #e0f2f1; /* Light Teal Background */
            border: 1px solid #00796b;
            padding: 20px;
            text-align: center;
            font-size: 1.1em;
        }

        /* Key CSS: Styling for the first item to make it span 2 columns */
        .item-1 {
            /* Starts at column line 1 and ends before column line 3, spanning 2 columns */
            grid-column: 1 / 3;
            
            /* Optional: For visual distinction */
            background-color: #4db6ac; /* Medium Teal */
            color: white;
            font-weight: bold;
        }
    </style>
</head>
<body>

    <div class="grid-container">
        <div class="grid-item item-1">Header/Item 1 (Spans 2 Columns)</div>
        <div class="grid-item item-2">Item 2 (Column 1 of Row 2)</div>
        <div class="grid-item item-3">Item 3 (Column 2 of Row 2)</div>
    </div>

</body>
</html>
```

Answer :Here is the required media query :


```
@media (max-width: 599px) {
    body {
        background-color: lightgrey;
    }
}
```

I am writing this inside the HTML code :



```
<!DOCTYPE html>
<html>
<head>
    <title>Media Query Example</title>
    <style>
        /* Default Styles for screens wider than 600px */
        body {
            background-color: #ffffff; /* White background by default */
            transition: background-color 0.3s ease;
            font-family: 'Inter', sans-serif;
            padding: 20px;
            min-height: 100vh;
            display: flex;
            justify-content: center;
            align-items: center;
            text-align: center;
        }

        .message-box {
            background-color: #f0f8ff;
            border-radius: 12px;
            padding: 30px;
            box-shadow: 0 4px 10px rgba(0, 0, 0, 0.1);
            max-width: 90%;
        }

        h1 {
            color: #1a73e8; /* Google Blue */
            margin-bottom: 15px;
        }

        p {
            color: #3c4043; /* Dark Grey */
            line-height: 1.6;
        }

        /* * Question 3 Answer: The Media Query
         * This applies styles when the screen's maximum width is 599px (i.e., less than 600px).
         */
        @media (max-width: 599px) {
            body {
                background-color: #d3d3d3; /* Light Grey background for small screens */
            }

            .message-box {
                background-color: #fffacd; /* Lemon Chiffon for contrast on small screens */
            }
        }
    </style>
</head>
<body>
    <div class="message-box">
        <h1>Resize Your Window!</h1>
        <p>This page demonstrates a CSS media query.</p>
        <p>If the browser width is **600 pixels or wider**, the background is **White**.</p>
        <p>If the browser width is **less than 600 pixels**, the background will change to **Light Grey**.</p>
        <p>Try reducing the width of your browser window or viewing this on a mobile device!</p>
    </div>
</body>
</html>

```

Answer :



```position: absolute;```  :
- Positioning Context: Relative to its closest non-statically positioned parent. If no such ancestor exists, it's positioned relative to the initial containing block (usually the `<html> `element).

- Scrolling Behavior: When the user scrolls the page, the element scrolls along with the rest of the content and is no longer visible once the viewport has scrolled past its position.



```position: fixed;```

- Positioning Context: Relative to the browser viewport (the visible area of the web page).

- Scrolling Behavior: When the user scrolls the page, the element remains in the same position on the screen. It is 'stuck' or 'fixed' in place, which is ideal for persistent navigation bars, sidebars, or 'back to top' buttons.
